```python
import os
from sklearn.model_selection import ParameterSampler
from tqdm.auto import tqdm

# Create the output directory
os.makedirs('stage_2_output', exist_ok=True)

# ── 9. HYPERPARAMETER RANDOM SEARCH ──────────────────────────

# Define parameter distributions to sample from
xgb_param_dist = {
    'n_estimators': [100, 200, 300, 400, 500, 600, 800, 1000],
    'max_depth': [3, 4, 5, 6, 8, 10],
    'learning_rate': [0.01, 0.03, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bylevel': [0.6, 0.7, 0.8, 0.9, 1.0],
    'min_child_weight': [1, 3, 5, 7, 10],
    'reg_lambda': [0.1, 1.0, 2.0, 5.0, 10.0],
    'reg_alpha': [0.0, 0.1, 0.5, 1.0, 5.0],
    'gamma': [0, 0.1, 0.2, 0.5, 1.0],
    'max_delta_step': [0, 1, 3, 5],
    'scale_pos_weight': [1, 2, 5, 10, scale_pos_weight, scale_pos_weight*2]
}

rf_param_dist = {
    'n_estimators': [100, 200, 300, 400, 500, 600, 800, 1000],
    'max_depth': [4, 6, 8, 10, None, 3],
    'class_weight': ['balanced', 'balanced_subsample', None],
    'min_samples_leaf': [1, 2, 3, 5],
    'max_features': ['sqrt', 'log2', None],
    'min_samples_split': [2, 5, 10, 15],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'bootstrap': [True, False]
}

# Number of random trials
n_trials = 3000

# Generate parameter samples
xgb_sampler = list(ParameterSampler(xgb_param_dist, n_iter=n_trials, random_state=42))
rf_sampler  = list(ParameterSampler(rf_param_dist, n_iter=n_trials, random_state=42))

xgb_trial_records = []
rf_trial_records = []

print(f"\n── Starting Hyperparameter Search ({n_trials} Trials) ──\n")

# Use tqdm for the progress bar, which automatically tracks iterations and estimates remaining time (ETA)
for trial_id in tqdm(range(n_trials), desc="Trials progress"):
    
    # -------- XGBoost --------
    xgb_params = xgb_sampler[trial_id]
    
    xgb_model_cv = xgb.XGBClassifier(
        **xgb_params,
        eval_metric           = 'aucpr',
        early_stopping_rounds = 30,
        random_state          = 42,
        verbosity             = 0
    )
    
    xgb_oof_cv = np.zeros(len(y))
    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y, groups=groups)):
        X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
        y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
        
        xgb_model_cv.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)
        xgb_oof_cv[val_idx] = xgb_model_cv.predict_proba(X_val)[:, 1]
    
    xgb_auc_cv = roc_auc_score(y, xgb_oof_cv)
    xgb_ap_cv  = average_precision_score(y, xgb_oof_cv)
    
    xgb_trial_records.append({
        'trial_id': trial_id + 1,
        'ap': xgb_ap_cv,
        'auc': xgb_auc_cv,
        **xgb_params
    })
    
    # Save XGBoost OOF array
    xgb_oof_df = gdf[['GLAKE_ID', 'glof_happened']].copy()
    xgb_oof_df['xgb_prob'] = xgb_oof_cv
    xgb_oof_df.to_csv(f'stage_2_output/oof_xgb_trial_{trial_id + 1}.csv', index=False)
    print(f"[XGBoost] AP: {xgb_ap_cv:.4f} | AUC: {xgb_auc_cv:.4f}")
    
    # -------- Random Forest --------
    rf_params = rf_sampler[trial_id]
    
    rf_model_cv = RandomForestClassifier(
        **rf_params,
        random_state = 42,
        n_jobs       = -1
    )
    
    rf_oof_cv = np.zeros(len(y))
    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y, groups=groups)):
        # Random Forest does not accept validation sets during fit out-of-the-box in the same way
        rf_model_cv.fit(X.iloc[train_idx], y.iloc[train_idx])
        rf_oof_cv[val_idx] = rf_model_cv.predict_proba(X.iloc[val_idx])[:, 1]
    
    rf_auc_cv = roc_auc_score(y, rf_oof_cv)
    rf_ap_cv  = average_precision_score(y, rf_oof_cv)
    
    rf_trial_records.append({
        'trial_id': trial_id + 1,
        'ap': rf_ap_cv,
        'auc': rf_auc_cv,
        **rf_params
    })
    
    # Save Random Forest OOF array
    rf_oof_df = gdf[['GLAKE_ID', 'glof_happened']].copy()
    rf_oof_df['rf_prob'] = rf_oof_cv
    rf_oof_df.to_csv(f'stage_2_output/oof_rf_trial_{trial_id + 1}.csv', index=False)
    print(f"[RF]      AP: {rf_ap_cv:.4f} | AUC: {rf_auc_cv:.4f}")
    print("-" * 35)

# -------- Save Top 5 Trials --------
print("\nSaving top 5 hyperparameter settings based on AP to separate CSVs...")

xgb_trials_df = pd.DataFrame(xgb_trial_records).sort_values('ap', ascending=False)
xgb_top_5 = xgb_trials_df.head(5)
xgb_top_5.to_csv('stage_2_output/top_5_xgb_hyperparameters.csv', index=False)

rf_trials_df = pd.DataFrame(rf_trial_records).sort_values('ap', ascending=False)
rf_top_5 = rf_trials_df.head(5)
rf_top_5.to_csv('stage_2_output/top_5_rf_hyperparameters.csv', index=False)

print("\n── Hyperparameter Search Completed successfully! ──")
print("Top 5 XGBoost Scores:")
print(xgb_top_5[['trial_id', 'ap', 'auc']].to_string(index=False))
print("\nTop 5 Random Forest Scores:")
print(rf_top_5[['trial_id', 'ap', 'auc']].to_string(index=False))
```

```python
import os
import joblib
import optuna
import numpy as np
import pandas as pd
import xgboost as xgb
import catboost as cb
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, average_precision_score

optuna.logging.set_verbosity(optuna.logging.WARNING)

# Create the new directories
os.makedirs('stage_2_output_1/best_models', exist_ok=True)

# Helper function to calculate recall at k
def recall_at_k(y_true, y_pred, k=200):
    sorted_idx = np.argsort(y_pred)[::-1]
    caught = y_true.iloc[sorted_idx[:k]].sum()
    n_pos = y_true.sum()
    return caught / n_pos

# Dictionaries to store out-of-fold predictions for every trial
xgb_oof_list = {}
rf_oof_list = {}
cb_oof_list = {}

# ── 1. XGBOOST OPTUNA OBJECTIVE ──────────────────────────────
def xgb_objective(trial):
    params = {
        'n_estimators':      trial.suggest_int('n_estimators', 100, 1000),
        'max_depth':         trial.suggest_int('max_depth', 3, 10),
        'learning_rate':     trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'subsample':         trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree':  trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'min_child_weight':  trial.suggest_int('min_child_weight', 1, 10),
        'reg_lambda':        trial.suggest_float('reg_lambda', 0.1, 10.0, log=True),
        'reg_alpha':         trial.suggest_float('reg_alpha', 0.0, 5.0),
        'gamma':             trial.suggest_float('gamma', 0.0, 5.0), # Expanded max to simulate focal loss via gamma
        'scale_pos_weight':  trial.suggest_float('scale_pos_weight', 1.0, scale_pos_weight * 5),
    }
    
    oof = np.zeros(len(y))
    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y, groups=groups)):
        # Model initialization moved safely INSIDE the loop to prevent cross-fold leakage
        model = xgb.XGBClassifier(
            **params,
            eval_metric           = 'aucpr',
            early_stopping_rounds = 30,
            random_state          = 42,
            verbosity             = 0
        )
        model.fit(
            X.iloc[train_idx], y.iloc[train_idx],
            eval_set = [(X.iloc[val_idx], y.iloc[val_idx])],
            verbose  = False
        )
        oof[val_idx] = model.predict_proba(X.iloc[val_idx])[:, 1]
    
    rec_200 = recall_at_k(y, oof, k=200)
    rec_100 = recall_at_k(y, oof, k=100)
    rec_50  = recall_at_k(y, oof, k=50)
    auc = roc_auc_score(y, oof)
    ap = average_precision_score(y, oof)
    
    trial.set_user_attr('recall_200', rec_200)
    trial.set_user_attr('recall_100', rec_100)
    trial.set_user_attr('recall_50', rec_50)
    trial.set_user_attr('auc', auc)
    trial.set_user_attr('ap', ap)
    
    # Combined objective: AP ensures learning is consistently guided by entire Precision-Recall space
    score = (0.7 * ap + 0.3 * auc)
    
    xgb_oof_list[trial.number] = oof
    return score

print("── Running Optuna for XGBoost (1000 Trials) ──")
xgb_study = optuna.create_study(direction='maximize')
xgb_study.optimize(xgb_objective, n_trials=1000, show_progress_bar=True)

print(f"XGBoost Best score: {xgb_study.best_value:.4f}")
best_xgb = xgb.XGBClassifier(**xgb_study.best_params, random_state=42, verbosity=0)
best_xgb.fit(X, y)
joblib.dump(best_xgb, 'stage_2_output_1/best_models/xgb_best_model.joblib')

xgb_metrics, xgb_params_list = [], []
for t in xgb_study.trials:
    if t.state.name == 'COMPLETE':
        xgb_metrics.append({
            'trial_id': t.number, 'ap': t.user_attrs.get('ap'), 'auc': t.user_attrs.get('auc'),
            'recall_200': t.user_attrs.get('recall_200'), 'recall_100': t.user_attrs.get('recall_100'),
            'recall_50': t.user_attrs.get('recall_50'), 'combined_score': t.value
        })
        p = {'trial_id': t.number}
        p.update(t.params)
        xgb_params_list.append(p)
pd.DataFrame(xgb_metrics).to_csv('stage_2_output_1/xgb_metrics_all_trials.csv', index=False)
pd.DataFrame(xgb_params_list).to_csv('stage_2_output_1/xgb_hyperparameters_all_trials.csv', index=False)

print("Saving Top 40 XGBoost OOF results...")
top_xgb_trials = sorted(xgb_study.trials, key=lambda t: t.value if t.value is not None else -1, reverse=True)[:40]
for t in top_xgb_trials:
    oof_results = gdf[feature_cols + ['GLAKE_ID', 'glof_happened', 'event_date']].copy()
    oof_results['xgb_prob'] = xgb_oof_list[t.number]
    oof_results.to_csv(f'stage_2_output_1/oof_xgb_{t.number:04d}.csv', index=False)


# ── 2. RANDOM FOREST OPTUNA OBJECTIVE ────────────────────────
def rf_objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_categorical('max_depth', [4, 6, 8, 10, None, 3]),
        'class_weight': trial.suggest_categorical('class_weight', ['balanced', 'balanced_subsample', None]),
        'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 5),
        'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None]),
        'min_samples_split': trial.suggest_int('min_samples_split', 2, 15),
        'criterion': trial.suggest_categorical('criterion', ['gini', 'entropy']),
        'bootstrap': trial.suggest_categorical('bootstrap', [True, False])
    }
    
    oof = np.zeros(len(y))
    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y, groups=groups)):
        model = RandomForestClassifier(**params, random_state=42, n_jobs=-1)
        model.fit(X.iloc[train_idx], y.iloc[train_idx])
        oof[val_idx] = model.predict_proba(X.iloc[val_idx])[:, 1]
        
    rec_200 = recall_at_k(y, oof, k=200)
    rec_100 = recall_at_k(y, oof, k=100)
    rec_50 = recall_at_k(y, oof, k=50)
    auc = roc_auc_score(y, oof)
    ap = average_precision_score(y, oof)
    
    trial.set_user_attr('recall_200', rec_200)
    trial.set_user_attr('recall_100', rec_100)
    trial.set_user_attr('recall_50', rec_50)
    trial.set_user_attr('auc', auc)
    trial.set_user_attr('ap', ap)
    
    score = (0.7 * ap + 0.3 * auc)
    
    rf_oof_list[trial.number] = oof
    return score

print("\n── Running Optuna for Random Forest (200 Trials) ──")
rf_study = optuna.create_study(direction='maximize')
rf_study.optimize(rf_objective, n_trials=200, show_progress_bar=True)

print(f"Random Forest Best score: {rf_study.best_value:.4f}")
best_rf = RandomForestClassifier(**rf_study.best_params, random_state=42, n_jobs=-1)
best_rf.fit(X, y)
joblib.dump(best_rf, 'stage_2_output_1/best_models/rf_best_model.joblib')

rf_metrics, rf_params_list = [], []
for t in rf_study.trials:
    if t.state.name == 'COMPLETE':
        rf_metrics.append({
            'trial_id': t.number, 'ap': t.user_attrs.get('ap'), 'auc': t.user_attrs.get('auc'),
            'recall_200': t.user_attrs.get('recall_200'), 'recall_100': t.user_attrs.get('recall_100'),
            'recall_50': t.user_attrs.get('recall_50'), 'combined_score': t.value
        })
        p = {'trial_id': t.number}
        p.update(t.params)
        rf_params_list.append(p)
pd.DataFrame(rf_metrics).to_csv('stage_2_output_1/rf_metrics_all_trials.csv', index=False)
pd.DataFrame(rf_params_list).to_csv('stage_2_output_1/rf_hyperparameters_all_trials.csv', index=False)

print("Saving Top 40 Random Forest OOF results...")
top_rf_trials = sorted(rf_study.trials, key=lambda t: t.value if t.value is not None else -1, reverse=True)[:40]
for t in top_rf_trials:
    oof_results = gdf[feature_cols + ['GLAKE_ID', 'glof_happened', 'event_date']].copy()
    oof_results['rf_prob'] = rf_oof_list[t.number]
    oof_results.to_csv(f'stage_2_output_1/oof_rf_{t.number:04d}.csv', index=False)


# ── 3. CATBOOST OPTUNA OBJECTIVE ──────────────────────────────
def cb_objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 200, 1000),
        'depth': trial.suggest_int('depth', 4, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1.0, 10.0),
        'auto_class_weights': trial.suggest_categorical('auto_class_weights', ['Balanced', 'SqrtBalanced', 'None']),
        'grow_policy': trial.suggest_categorical('grow_policy', ['SymmetricTree', 'Depthwise']),
    }
    
    oof = np.zeros(len(y))
    for fold, (train_idx, val_idx) in enumerate(cv.split(X, y, groups=groups)):
        model = cb.CatBoostClassifier(**params, random_state=42, verbose=0)
        model.fit(
            X.iloc[train_idx], y.iloc[train_idx],
            eval_set=(X.iloc[val_idx], y.iloc[val_idx]),
            early_stopping_rounds=30,
            verbose=False
        )
        oof[val_idx] = model.predict_proba(X.iloc[val_idx])[:, 1]
        
    rec_200 = recall_at_k(y, oof, k=200)
    rec_100 = recall_at_k(y, oof, k=100)
    rec_50  = recall_at_k(y, oof, k=50)
    auc = roc_auc_score(y, oof)
    ap = average_precision_score(y, oof)
    
    trial.set_user_attr('recall_200', rec_200)
    trial.set_user_attr('recall_100', rec_100)
    trial.set_user_attr('recall_50', rec_50)
    trial.set_user_attr('auc', auc)
    trial.set_user_attr('ap', ap)
    
    # Combined objective
    score = (0.7 * ap + 0.3 * auc)
    
    cb_oof_list[trial.number] = oof
    return score

print("\n── Running Optuna for CatBoost (500 Trials) ──")
cb_study = optuna.create_study(direction='maximize')
cb_study.optimize(cb_objective, n_trials=500, show_progress_bar=True)

print(f"CatBoost Best score: {cb_study.best_value:.4f}")
best_cb = cb.CatBoostClassifier(**cb_study.best_params, random_state=42, verbose=0)
best_cb.fit(X, y)
joblib.dump(best_cb, 'stage_2_output_1/best_models/cb_best_model.joblib')

cb_metrics, cb_params_list = [], []
for t in cb_study.trials:
    if t.state.name == 'COMPLETE':
        cb_metrics.append({
            'trial_id': t.number, 'ap': t.user_attrs.get('ap'), 'auc': t.user_attrs.get('auc'),
            'recall_200': t.user_attrs.get('recall_200'), 'recall_100': t.user_attrs.get('recall_100'),
            'recall_50': t.user_attrs.get('recall_50'), 'combined_score': t.value
        })
        p = {'trial_id': t.number}
        p.update(t.params)
        cb_params_list.append(p)
pd.DataFrame(cb_metrics).to_csv('stage_2_output_1/cb_metrics_all_trials.csv', index=False)
pd.DataFrame(cb_params_list).to_csv('stage_2_output_1/cb_hyperparameters_all_trials.csv', index=False)

print("Saving Top 40 CatBoost OOF results...")
top_cb_trials = sorted(cb_study.trials, key=lambda t: t.value if t.value is not None else -1, reverse=True)[:40]
for t in top_cb_trials:
    oof_results = gdf[feature_cols + ['GLAKE_ID', 'glof_happened', 'event_date']].copy()
    oof_results['cb_prob'] = cb_oof_list[t.number]
    oof_results.to_csv(f'stage_2_output_1/oof_cb_{t.number:04d}.csv', index=False)

print("\n── Process Complete! Best models, metrics, and Top 40 predictions are saved for XGB, RF, and CB. ──")
```

```python
import os
import copy
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, WeightedRandomSampler
from sklearn.preprocessing import QuantileTransformer
from sklearn.impute import SimpleImputer
import optuna
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score, average_precision_score

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")

# Preprocess dataset for PyTorch (Imputation to handle NaNs missing in XGB/RF sets)
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# ── 1. HELPERS & LOSSES ────────────────────────────
def get_activation(name: str):
    if name == "relu": return nn.ReLU()
    if name == "leakyrelu": return nn.LeakyReLU(negative_slope=0.1)
    if name == "gelu": return nn.GELU()
    raise ValueError(f"Unknown activation: {name}")

class FocalLossWithLogits(nn.Module):
    def __init__(self, alpha=0.25, gamma=2.0, reduction="mean"):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction

    def forward(self, logits, targets):
        targets = targets.float()
        bce = nn.functional.binary_cross_entropy_with_logits(logits, targets, reduction="none")
        p = torch.sigmoid(logits)
        p_t = p * targets + (1 - p) * (1 - targets)
        alpha_t = self.alpha * targets + (1 - self.alpha) * (1 - targets)
        loss = alpha_t * ((1 - p_t) ** self.gamma) * bce
        if self.reduction == "mean": return loss.mean()
        if self.reduction == "sum": return loss.sum()
        return loss

def pairwise_ranking_loss(probs, labels, margin=0.1, neg_multiplier=20):
    pos_mask = labels == 1
    neg_mask = labels == 0
    pos_probs = probs[pos_mask]
    neg_probs = probs[neg_mask]
    if pos_probs.numel() == 0 or neg_probs.numel() == 0:
        return torch.zeros((), device=probs.device, dtype=probs.dtype)
    n_sample = min(int(pos_probs.numel() * neg_multiplier), int(neg_probs.numel()))
    idx = torch.randperm(neg_probs.numel(), device=probs.device)[:n_sample]
    neg_sample = neg_probs[idx]
    diff = pos_probs.unsqueeze(1) - neg_sample.unsqueeze(0) - margin
    return torch.clamp(-diff, min=0).mean()

# ── 2. PYTORCH ARCHITECTURES ────────────────────────────
def decreasing_sizes(hidden: int, num_blocks: int):
    return [max(8, hidden // (2 ** i)) for i in range(num_blocks)]

def incdec_sizes(hidden: int, num_blocks: int):
    h2 = max(8, hidden // 2)
    h4 = max(8, hidden // 4)
    if num_blocks == 3: return [h2, hidden, h2]
    if num_blocks == 4: return [h4, h2, hidden, h2]
    if num_blocks == 5: return [h4, h2, hidden, h2, h4]
    return [h2, hidden, h2]

class NormalMLPDecreasing(nn.Module):
    def __init__(self, in_features: int, hidden: int, num_blocks: int, dropout: float, activation: str):
        super().__init__()
        sizes = decreasing_sizes(hidden, num_blocks)
        layers = []
        prev = in_features
        for s in sizes:
            layers.extend([
                nn.Linear(prev, s), get_activation(activation),
                nn.BatchNorm1d(s), nn.Dropout(dropout)
            ])
            prev = s
        layers.append(nn.Linear(prev, 1))
        self.net = nn.Sequential(*layers)
    def forward(self, x): return self.net(x).squeeze(1)

class NormalMLPIncDec(nn.Module):
    def __init__(self, in_features: int, hidden: int, num_blocks: int, dropout: float, activation: str):
        super().__init__()
        sizes = incdec_sizes(hidden, num_blocks)
        layers = []
        prev = in_features
        for s in sizes:
            layers.extend([
                nn.Linear(prev, s), get_activation(activation),
                nn.BatchNorm1d(s), nn.Dropout(dropout)
            ])
            prev = s
        layers.append(nn.Linear(prev, 1))
        self.net = nn.Sequential(*layers)
    def forward(self, x): return self.net(x).squeeze(1)

class ResidualMLP(nn.Module):
    def __init__(self, in_features: int, hidden: int, num_blocks: int, dropout: float, activation: str):
        super().__init__()
        sizes = decreasing_sizes(hidden, num_blocks)
        self.blocks = nn.ModuleList()
        prev = in_features
        for s in sizes:
            self.blocks.append(nn.ModuleDict({
                "fc": nn.Linear(prev, s),
                "act": get_activation(activation),
                "bn": nn.BatchNorm1d(s),
                "drop": nn.Dropout(dropout),
            }))
            prev = s
        self.out = nn.Linear(prev, 1)

    def forward(self, x):
        for block in self.blocks:
            x_prev = x
            x = block["fc"](x)
            x = block["act"](x)
            x = block["bn"](x)
            x = block["drop"](x)
            if x_prev.shape[-1] == x.shape[-1]:
                x = x + x_prev
        return self.out(x).squeeze(1)

def build_model(model_type: str, in_features: int, hidden: int, num_blocks: int, dropout: float, activation: str):
    if model_type == "residual_mlp": return ResidualMLP(in_features, hidden, num_blocks, dropout, activation)
    if model_type == "normal_mlp_decreasing": return NormalMLPDecreasing(in_features, hidden, num_blocks, dropout, activation)
    if model_type == "normal_mlp_incdec": return NormalMLPIncDec(in_features, hidden, num_blocks, dropout, activation)
    raise ValueError(f"Unknown model_type: {model_type}")

torch_oof_list = {}

def topk_capture(y_true, probs, k=500):
    n = len(y_true)
    if n == 0:
        return 0, 0, 0.0
    kk = min(k, n)
    ranked_idx = np.argsort(-probs)
    top_idx = ranked_idx[:kk]
    hits = int((y_true[top_idx] == 1).sum())
    total_pos = int((y_true == 1).sum())
    recall_k = hits / max(1, total_pos)
    return hits, total_pos, float(recall_k)

# ── 3. OPTUNA OBJECTIVE ───────────────────────────────
def torch_objective(trial):
    # Hyperparameters based on model.ipynb setup
    model_type = trial.suggest_categorical('model_type', ["residual_mlp", "normal_mlp_decreasing", "normal_mlp_incdec"])
    hidden = trial.suggest_categorical('hidden_size', [64,128, 256, 384, 512])
    num_blocks = trial.suggest_int('num_blocks', 3, 7)
    dropout = trial.suggest_categorical('dropout', [0.2, 0.3, 0.4])
    activation = trial.suggest_categorical('activation', ["relu", "leakyrelu", "gelu"])
    
    alpha = trial.suggest_categorical('focal_alpha', [0.25, 0.5, 0.75])
    gamma = trial.suggest_categorical('focal_gamma', [1.0, 2.0, 3.0])
    pos_amp = trial.suggest_categorical('pos_weight_amp', [1, 2, 5, 10, 20])
    
    # Loss weights summing to ~1
    bce_weight = trial.suggest_float('bce_weight', 0.2,1)
    focal_weight = trial.suggest_float('focal_weight', 0.1,1)
    rank_weight = 1.0 - (bce_weight + focal_weight) 
    
    rank_margin = trial.suggest_categorical('rank_margin', [0.05, 0.10])
    rank_neg_mult = trial.suggest_categorical('rank_neg_multiplier', [20, 30])
    
    lr = trial.suggest_float('lr', 1e-4, 5e-3, log=True)
    weight_decay = trial.suggest_float('weight_decay', 1e-5, 1e-3, log=True)
    batch_size = 128
    epochs = 350 # Using patience/early stopping
    
    oof = np.zeros(len(y))
    fold_aps = []
    
    for fold, (train_idx, val_idx) in enumerate(cv.split(X_imputed, y, groups=groups)):
        X_tr_df = X_imputed.iloc[train_idx]
        X_va_df = X_imputed.iloc[val_idx]
        y_tr = y.iloc[train_idx].values
        y_va = y.iloc[val_idx].values
        
        # QuantileTransformer
        prep = QuantileTransformer(n_quantiles=min(1000, len(train_idx)), output_distribution="uniform", random_state=42)
        X_tr = prep.fit_transform(X_tr_df.values).astype(np.float32)
        X_va = prep.transform(X_va_df.values).astype(np.float32)
        
        X_tr_t = torch.tensor(X_tr)
        y_tr_t = torch.tensor(y_tr, dtype=torch.float32)
        X_va_t = torch.tensor(X_va).to(DEVICE)
        
        # Weighted sampler
        class_counts = np.bincount(y_tr)
        class_weights = 1.0 / np.maximum(class_counts, 1)
        sample_weights = class_weights[y_tr]
        sampler = WeightedRandomSampler(weights=torch.tensor(sample_weights, dtype=torch.double), num_samples=len(sample_weights), replacement=True)
        
        train_loader = DataLoader(TensorDataset(X_tr_t, y_tr_t), batch_size=batch_size, sampler=sampler, drop_last=False)
        
        model = build_model(model_type, X_tr.shape[1], hidden, num_blocks, dropout, activation).to(DEVICE)
        optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode="max", factor=0.5, patience=8, min_lr=1e-5)
        
        pos_weight_value = float((n_neg / max(1, n_pos)) * pos_amp)
        bce_criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([pos_weight_value], dtype=torch.float32, device=DEVICE))
        focal_criterion = FocalLossWithLogits(alpha=alpha, gamma=gamma)
        
        best_fold_score = -1.0
        best_state = None
        bad_epochs = 0
        patience = 25
        
        for epoch in range(epochs):
            model.train()
            for xb, yb in train_loader:
                xb, yb = xb.to(DEVICE), yb.to(DEVICE)
                optimizer.zero_grad()
                logits = model(xb)
                probs = torch.sigmoid(logits)
                
                loss_bce = bce_criterion(logits, yb)
                loss_focal = focal_criterion(logits, yb)
                loss_rank = pairwise_ranking_loss(probs, yb, margin=rank_margin, neg_multiplier=rank_neg_mult)
                
                loss = (bce_weight * loss_bce) + (focal_weight * loss_focal) + (rank_weight * loss_rank)
                loss.backward()
                optimizer.step()
                
            model.eval()
            with torch.no_grad():
                val_logits = model(X_va_t)
                val_probs_epoch = torch.sigmoid(val_logits).cpu().numpy()
                
            # Early stopping metric (matches model.ipynb logic)
            val_ap = float(average_precision_score(y_va, val_probs_epoch))
            _, _, val_top500_rec = topk_capture(y_va, val_probs_epoch, k=500)
            
            val_goal = 0.60 * val_top500_rec + 0.40 * val_ap
            scheduler.step(val_goal)
            
            if val_goal > best_fold_score + 1e-5:
                best_fold_score = val_goal
                best_state = copy.deepcopy(model.state_dict())
                bad_epochs = 0
            else:
                bad_epochs += 1
                if bad_epochs >= patience:
                    break
                    
        # Load best and predict
        if best_state: model.load_state_dict(best_state)
        model.eval()
        with torch.no_grad():
            val_logits = model(X_va_t)
            val_probs = torch.sigmoid(val_logits).cpu().numpy()
            
        oof[val_idx] = val_probs
        fold_aps.append(average_precision_score(y_va, val_probs))
        
    rec_200 = topk_capture(y, oof, k=200)[2]
    rec_100 = topk_capture(y, oof, k=100)[2]
    rec_50  = topk_capture(y, oof, k=50)[2]
    auc = roc_auc_score(y, oof)
    ap = average_precision_score(y, oof)
    mean_fold_ap = np.mean(fold_aps)
    
    trial.set_user_attr('recall_200', rec_200)
    trial.set_user_attr('recall_100', rec_100)
    trial.set_user_attr('recall_50', rec_50)
    trial.set_user_attr('auc', auc)
    trial.set_user_attr('ap', ap)
    trial.set_user_attr('mean_fold_ap', mean_fold_ap)
    
    # Matching custom goal_score from original script
    # (0.45 * top500_score + 0.25 * mean_fold_ap + 0.15 * oof_ap) (simplified for Optuna)
    rec_500 = topk_capture(y, oof, k=500)[2]
    score = (0.50 * rec_500 + 0.30 * mean_fold_ap + 0.20 * ap)
    
    torch_oof_list[trial.number] = oof
    return score

print("\n── Running Optuna for PyTorch 3-Arch (150 Trials) ──")
torch_study = optuna.create_study(direction='maximize')
torch_study.optimize(torch_objective, n_trials=2000, show_progress_bar=True)

print(f"PyTorch Best score: {torch_study.best_value:.4f}")
print(f"PyTorch Best params: {torch_study.best_params}")

# ── 4. SAVE METRICS, OOF & BEST MODEL ───────────────────────────
os.makedirs('stage_2_pytorch', exist_ok=True)

torch_metrics, torch_params_list = [], []
for t in torch_study.trials:
    if t.state.name == 'COMPLETE':
        torch_metrics.append({
            'trial_id': t.number, 'ap': t.user_attrs.get('ap'), 'auc': t.user_attrs.get('auc'),
            'mean_fold_ap': t.user_attrs.get('mean_fold_ap'),
            'recall_200': t.user_attrs.get('recall_200'), 'recall_100': t.user_attrs.get('recall_100'),
            'recall_50': t.user_attrs.get('recall_50'), 'combined_score': t.value
        })
        p = {'trial_id': t.number}
        p.update(t.params)
        torch_params_list.append(p)
        
pd.DataFrame(torch_metrics).to_csv('stage_2_pytorch/torch_metrics_all_trials.csv', index=False)
pd.DataFrame(torch_params_list).to_csv('stage_2_pytorch/torch_hyperparameters_all_trials.csv', index=False)

print("Saving Top 20 PyTorch OOF results...")
top_torch_trials = sorted(torch_study.trials, key=lambda t: t.value if t.value is not None else -1, reverse=True)[:20]
for t in top_torch_trials:
    arch_name = t.params.get('model_type', 'mlp')
    oof_results = gdf[feature_cols + ['GLAKE_ID', 'glof_happened', 'event_date','geometry']].copy()
    oof_results['torch_prob'] = torch_oof_list[t.number]
    oof_results.to_csv(f'stage_2_pytorch/oof_{arch_name}_{t.number:04d}.csv', index=False)

print("Training and Saving Best PyTorch Model on Full Data...")
best = torch_study.best_params
best_model = build_model(
    model_type=best['model_type'], 
    in_features=X_imputed.shape[1], 
    hidden=best['hidden_size'], 
    num_blocks=best['num_blocks'], 
    dropout=best['dropout'], 
    activation=best['activation']
).to(DEVICE)

# Brief full-data training for the best model instance
prep = QuantileTransformer(n_quantiles=min(1000, len(X_imputed)), output_distribution="uniform", random_state=42)
X_full = prep.fit_transform(X_imputed.values).astype(np.float32)
X_full_t = torch.tensor(X_full)
y_full_t = torch.tensor(y.values, dtype=torch.float32)

class_counts = np.bincount(y.values)
class_weights = 1.0 / np.maximum(class_counts, 1)
sample_weights = class_weights[y.values]
sampler = WeightedRandomSampler(weights=torch.tensor(sample_weights, dtype=torch.double), num_samples=len(sample_weights), replacement=True)
full_loader = DataLoader(TensorDataset(X_full_t, y_full_t), batch_size=128, sampler=sampler)

bce_weight = best.get('bce_weight', 0.5)
focal_weight = best.get('focal_weight', 0.3)
rank_weight = 1.0 - (bce_weight + focal_weight)
pos_weight_value = float((n_neg / max(1, n_pos)) * best.get('pos_weight_amp', 5))
bce_criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([pos_weight_value], dtype=torch.float32, device=DEVICE))
focal_criterion = FocalLossWithLogits(alpha=best.get('focal_alpha', 0.25), gamma=best.get('focal_gamma', 2.0))

optimizer = torch.optim.AdamW(best_model.parameters(), lr=best.get('lr', 1e-3), weight_decay=best.get('weight_decay', 1e-4))

best_model.train()
for epoch in range(100): # fixed epochs for full training
    for xb, yb in full_loader:
        xb, yb = xb.to(DEVICE), yb.to(DEVICE)
        optimizer.zero_grad()
        logits = best_model(xb)
        probs = torch.sigmoid(logits)
        
        loss_bce = bce_criterion(logits, yb)
        loss_focal = focal_criterion(logits, yb)
        loss_rank = pairwise_ranking_loss(probs, yb, margin=best.get('rank_margin', 0.1), neg_multiplier=best.get('rank_neg_multiplier', 20))
        
        loss = (bce_weight * loss_bce) + (focal_weight * loss_focal) + (rank_weight * loss_rank)
        loss.backward()
        optimizer.step()

# Save the trained model state
torch.save(best_model.state_dict(), 'stage_2_pytorch/best_torch_model.pth')

print("\n── PyTorch Optuna Search Complete! ──")
```
