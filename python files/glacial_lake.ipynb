```python
import geopandas as gpd
import pandas as pd

# --- Your loaded layers ---
# gdf_2018  → has a glacial lake ID column, let's say it's called "gl_id"
# gdf_2020  → no ID yet

ID_COL = "GLAKE_ID"   # ← change to your actual column name in the 2018 layer
TARGET_CRS = "ESRI:102025"

# ── Step 1: Reproject both to equal-area ──────────────────────────────────
gdf_2018 = gdf_2018.to_crs(TARGET_CRS).reset_index(drop=True)
gdf_2020 = gdf_2020.to_crs(TARGET_CRS).reset_index(drop=True)

# Assign a temporary integer index to 2020 for tracking
gdf_2020["_idx_2020"] = gdf_2020.index
gdf_2018["_idx_2018"] = gdf_2018.index

# ── Step 2: Compute intersections ─────────────────────────────────────────
inter = gpd.overlay(
    gdf_2020[["_idx_2020", "geometry"]],
    gdf_2018[[ID_COL, "_idx_2018", "geometry"]],
    how="intersection",
    keep_geom_type=False
)

inter["overlap_area_m2"] = inter.geometry.area

# ── Step 3: For each 2020 polygon, keep the 2018 match with max overlap ───
best_match = (
    inter
    .sort_values("overlap_area_m2", ascending=False)
    .drop_duplicates(subset="_idx_2020", keep="first")   # one best match per 2020 lake
)

# ── Step 4: Overlap ratio filter (avoid spurious edge touches) ────────────
area_2020 = gdf_2020.set_index("_idx_2020").geometry.area
best_match["ref_area_m2"]   = best_match["_idx_2020"].map(area_2020)
best_match["overlap_ratio"] = best_match["overlap_area_m2"] / best_match["ref_area_m2"]

# Keep matches where at least 10% of the 2020 lake is covered by the 2018 polygon
# (stricter than before since these are close in time — only 2 years apart)
best_match = best_match[best_match["overlap_ratio"] >= 0.10]

print(f"2020 lakes total         : {len(gdf_2020)}")
print(f"Successfully matched     : {best_match['_idx_2020'].nunique()}")
print(f"Unmatched (new lakes?)   : {len(gdf_2020) - best_match['_idx_2020'].nunique()}")

# ── Step 5: Transfer the ID back to the 2020 layer ────────────────────────
id_map = best_match.set_index("_idx_2020")[ID_COL]

gdf_2020[ID_COL] = gdf_2020["_idx_2020"].map(id_map)
gdf_2020 = gdf_2020.drop(columns="_idx_2020")

# Flag unmatched lakes clearly
gdf_2020["id_source"] = gdf_2020[ID_COL].apply(
    lambda x: "matched_2018" if pd.notna(x) else "new_unmatched"
)

print(gdf_2020[ID_COL].isna().sum(), "lakes have no 2018 match (NaN id)")
print(gdf_2020.head())

# Save
gdf_2020.to_file("glacial_lakes_2020_with_id.gpkg", driver="GPKG")
```

```python
overlap_threshold = 0.10
years_hist = [1990,2000,2010,2015]

gdfs_hist = {
  1990:gdf_1990,
  2000:gdf_2000,
  2010:gdf_2010,
  2015:gdf_2015,
}

ref = gdf_2020.to_crs(TARGET_CRS).copy().reset_index(drop= True)
ref['_idx_ref'] = ref.index
ref['area_2020'] = ref.geometry.area
ref['perimeter_2020'] = ref.geometry.length
ref['geometry_2020'] = ref.geometry

master = ref.copy()
```

```python
def match(ref_gdf,hist_gdf, year, threshold):
  hist = hist_gdf.to_crs(TARGET_CRS).copy().reset_index(drop = True)
  hist['_idx_hist'] = hist.index

  inter = gpd.overlay(
    ref_gdf[['_idx_ref','geometry']],
    hist[['_idx_hist','geometry']],
    how='intersection',
    keep_geom_type=False
  )

  if inter.empty:
    print(f'no overlaps found for {year}')
    return pd.DataFrame(columns=["_idx_ref",
                                     f"area_{year}",
                                     f"perimeter_{year}",
                                     f"geometry_{year}",
                                     f"flag_{year}"])
  

  inter['overlap_area_m2'] = inter.geometry.area
  best = (inter
          .sort_values('overlap_area_m2',ascending=False)
          .drop_duplicates(subset='_idx_ref',keep='first')
          .copy())
  
  ref_area_map = ref_gdf.set_index('_idx_ref')['area_2020']
  best['ref_area_m2'] = best['_idx_ref'].map(ref_area_map)
  best["overlap_ratio"] = best["overlap_area_m2"] / best["ref_area_m2"]
  best = best[best["overlap_ratio"] >= threshold].copy()

  hist['area_m2'] = hist.geometry.area
  hist['perim_m'] = hist.geometry.length
  hist['geom'] = hist.geometry

  hist_info = hist.set_index('_idx_hist')[['area_m2','perim_m','geom']]

  best[f'area_{year}'] = best['_idx_hist'].map(hist_info['area_m2'])
  best[f'perimeter_{year}'] = best['_idx_hist'].map(hist_info['perim_m'])
  best[f'geometry_{year}'] = best['_idx_hist'].map(hist_info['geom'])
  best[f'flag_{year}'] = 'observed'

  print(f' [{year}] matched: {len(best)} / {len(ref_gdf)}')

  return best[["_idx_ref",
                 f"area_{year}",
                 f"perimeter_{year}",
                 f"geometry_{year}",
                 f"flag_{year}"]].reset_index(drop=True)
```

```python
for yr in years_hist:
  print(f'\nMatching {yr} -> 2020 ..')
  result = match(master, gdfs_hist[yr],yr,threshold = 0.10)


  if result.empty:
    master[f'area_{yr}'] = np.nan
    master[f'perimeter_{yr}'] = np.nan
    master[f'geometry_{yr}'] = np.nan
    master[f'flag_{yr}'] = 'not_mapped'
    continue

  master = master.merge(result, on= '_idx_ref',how = 'left')

```

```python
all_years = [1990, 2000, 2010, 2015, 2020]

def first_observed_year(row):
    for yr in all_years:
        if yr == 2020:
            return 2020   # always observed in reference
        if pd.notna(row.get(f"area_{yr}")):
            return yr
    return 2020

master["first_observed_yr"] = master.apply(first_observed_year, axis=1)

# Now fill flags properly
for yr in years_hist:
    flag_col = f"flag_{yr}"
    area_col = f"area_{yr}"

    def fill_flag(row):
        if pd.notna(row[area_col]):
            return "observed"
        # NaN → decide why
        if row["first_observed_yr"] > yr:
            return "not_formed"    # lake formed after this year
        else:
            return "not_mapped"    # lake existed but missing from shapefile

    master[flag_col] = master.apply(fill_flag, axis=1)


# ── Step 5: Sanity check ──────────────────────────────────────────────────
print("\n── Flag summary ──")
for yr in years_hist:
    print(f"\n{yr}:")
    print(master[f"flag_{yr}"].value_counts())


# ── Step 6: Convert areas to km² ─────────────────────────────────────────
for yr in all_years:
    master[f"area_{yr}_km2"] = master[f"area_{yr}"] / 1e6


# ── Step 7: Compute expansion rate (observed points only) ─────────────────
def expansion_rate(row):
    obs_yrs   = []
    obs_areas = []

    for yr in all_years:
        flag = row.get(f"flag_{yr}", "observed")   # 2020 has no flag
        area = row.get(f"area_{yr}_km2", np.nan)
        if pd.notna(area) and flag == "observed":
            obs_yrs.append(yr)
            obs_areas.append(area)

    if len(obs_areas) < 2:
        return pd.Series({
            "expansion_rate_km2_yr" : np.nan,
            "expansion_rate_pct_yr" : np.nan,
            "n_obs_points"          : len(obs_areas)
        })

    slope, _ = np.polyfit(obs_yrs, obs_areas, 1)
    norm_rate = slope / obs_areas[0] * 100

    return pd.Series({
        "expansion_rate_km2_yr" : round(slope, 8),
        "expansion_rate_pct_yr" : round(norm_rate, 4),
        "n_obs_points"          : len(obs_areas)
    })

master[["expansion_rate_km2_yr",
        "expansion_rate_pct_yr",
        "n_obs_points"]] = master.apply(expansion_rate, axis=1)

# ── Step 8: Convert extra geometry columns to WKT before saving ───────────
output = gpd.GeoDataFrame(master, geometry="geometry_2020", crs=TARGET_CRS)
output = output.drop(columns=["_idx_ref"])

# Identify all extra geometry columns (all years except the active 2020)
```

```python
from shapely.geometry.base import BaseGeometry

# Find ALL geometry-type columns except the active one
active_geom = output.geometry.name   # "geometry_2020"

def is_geom_column(series):
    """Check if a column contains geometry objects"""
    non_null = series.dropna()
    if len(non_null) == 0:
        return False
    return isinstance(non_null.iloc[0], BaseGeometry)

geom_cols_to_convert = [
    col for col in output.columns
    if col != active_geom and is_geom_column(output[col])
]

print(f"Converting these geometry columns to WKT: {geom_cols_to_convert}")

for col in geom_cols_to_convert:
    output[col] = gpd.GeoSeries(output[col], crs=TARGET_CRS).to_wkt()

# Verify no extra geometry columns remain
remaining = [col for col in output.columns
             if col != active_geom and is_geom_column(output[col])]
print(f"Remaining geometry columns after conversion: {remaining}")  # should be []

output.to_file("glacial_lakes_timeseries.gpkg", driver="GPKG")
print(f"Saved: {len(output)} lakes")
```

```python
gl_1990 = gpd.read_file('1. glacial lake datasets\glacial_lakes_1990_v1.shp')
gl_2015 = gpd.read_file('1. glacial lake datasets\glacial_lakes_2015_v1.shp')
```

```python
# Timeseries analysis: gl_1990 and 5gl_2015 → combined GeoPackage
# (Similar workflow to glacial_lakes_timeseries.gpkg)

TARGET_CRS_GL = "ESRI:102025"

# ── Step 1: Use 2018 as reference, reproject both to equal-area ──────────
ref_gl = gl_2015.to_crs(TARGET_CRS_GL).reset_index(drop=True).copy()
ref_gl["_idx_ref"] = ref_gl.index
ref_gl["area_2015"] = ref_gl.geometry.area
ref_gl["perimeter_2015"] = ref_gl.geometry.length
ref_gl["geometry_2015"] = ref_gl.geometry

master_gl = ref_gl.copy()

# ── Step 2: Match 1990 to 2018 ────────────────────────────────────────────
hist_gl_1990 = gl_1990.to_crs(TARGET_CRS_GL).copy().reset_index(drop=True)
hist_gl_1990["_idx_hist"] = hist_gl_1990.index

inter_gl = gpd.overlay(
    ref_gl[["_idx_ref", "geometry"]],
    hist_gl_1990[["_idx_hist", "geometry"]],
    how="intersection",
    keep_geom_type=False
)

overlap_threshold = 0.10

if not inter_gl.empty:
    inter_gl["overlap_area_m2"] = inter_gl.geometry.area
    best_gl = (
        inter_gl
        .sort_values("overlap_area_m2", ascending=False)
        .drop_duplicates(subset="_idx_ref", keep="first")
        .copy()
    )
    
    ref_area_map_gl = ref_gl.set_index("_idx_ref")["area_2015"]
    best_gl["ref_area_m2"] = best_gl["_idx_ref"].map(ref_area_map_gl)
    best_gl["overlap_ratio"] = best_gl["overlap_area_m2"] / best_gl["ref_area_m2"]
    best_gl = best_gl[best_gl["overlap_ratio"] >= overlap_threshold].copy()
    
    hist_gl_1990["area_m2"] = hist_gl_1990.geometry.area
    hist_gl_1990["perim_m"] = hist_gl_1990.geometry.length
    hist_gl_1990["geom"] = hist_gl_1990.geometry
    
    hist_info_gl = hist_gl_1990.set_index("_idx_hist")[["area_m2", "perim_m", "geom"]]
    
    best_gl["area_1990"] = best_gl["_idx_hist"].map(hist_info_gl["area_m2"])
    best_gl["perimeter_1990"] = best_gl["_idx_hist"].map(hist_info_gl["perim_m"])
    best_gl["geometry_1990"] = best_gl["_idx_hist"].map(hist_info_gl["geom"])
    best_gl["flag_1990"] = "observed"
    
    result_gl = best_gl[
        ["_idx_ref", "area_1990", "perimeter_1990", "geometry_1990", "flag_1990"]
    ].reset_index(drop=True)
    
    master_gl = master_gl.merge(result_gl, on="_idx_ref", how="left")
    print(f"[1990] matched: {len(result_gl)} / {len(master_gl)}")
else:
    print("No overlaps found for 1990")
    master_gl["area_1990"] = np.nan
    master_gl["perimeter_1990"] = np.nan
    master_gl["geometry_1990"] = np.nan
    master_gl["flag_1990"] = "not_mapped"

# ── Step 3: Fill flags ────────────────────────────────────────────────────
all_years_gl = [1990, 2015]

def first_observed_year_gl(row):
    for yr in all_years_gl:
        if yr == 2015:
            return 2015
        if pd.notna(row.get(f"area_{yr}")):
            return yr
    return 2015

master_gl["first_observed_yr"] = master_gl.apply(first_observed_year_gl, axis=1)

flag_col = "flag_1990"
area_col = "area_1990"

def fill_flag_gl(row):
    if pd.notna(row[area_col]):
        return "observed"
    if row["first_observed_yr"] > 1990:
        return "not_formed"
    else:
        return "not_mapped"

master_gl[flag_col] = master_gl.apply(fill_flag_gl, axis=1)

print("\n── Flag summary ──")
print(f"1990: {master_gl['flag_1990'].value_counts().to_dict()}")

# ── Step 4: Convert to km² ────────────────────────────────────────────────
for yr in all_years_gl:
    master_gl[f"area_{yr}_km2"] = master_gl[f"area_{yr}"] / 1e6

# ── Step 5: Compute expansion rate ────────────────────────────────────────
def expansion_rate_gl(row):
    obs_yrs = []
    obs_areas = []
    
    for yr in all_years_gl:
        flag = row.get(f"flag_{yr}", "observed")
        area = row.get(f"area_{yr}_km2", np.nan)
        if pd.notna(area) and flag == "observed":
            obs_yrs.append(yr)
            obs_areas.append(area)
    
    if len(obs_areas) < 2:
        return pd.Series({
            "expansion_rate_km2_yr": np.nan,
            "expansion_rate_pct_yr": np.nan,
            "n_obs_points": len(obs_areas)
        })
    
    slope, _ = np.polyfit(obs_yrs, obs_areas, 1)
    norm_rate = slope / obs_areas[0] * 100
    
    return pd.Series({
        "expansion_rate_km2_yr": round(slope, 8),
        "expansion_rate_pct_yr": round(norm_rate, 4),
        "n_obs_points": len(obs_areas)
    })

master_gl[["expansion_rate_km2_yr", "expansion_rate_pct_yr", "n_obs_points"]] = master_gl.apply(
    expansion_rate_gl, axis=1
)

# ── Step 6: Convert extra geometry columns to WKT ─────────────────────────
output_gl = gpd.GeoDataFrame(master_gl, geometry="geometry_2015", crs=TARGET_CRS_GL)
output_gl = output_gl.drop(columns=["_idx_ref"])

# Find all geometry columns to convert
active_geom_gl = output_gl.geometry.name

geom_cols_to_convert_gl = [
    col for col in output_gl.columns
    if col != active_geom_gl and is_geom_column(output_gl[col])
]

print(f"Converting geometry columns to WKT: {geom_cols_to_convert_gl}")

for col in geom_cols_to_convert_gl:
    output_gl[col] = gpd.GeoSeries(output_gl[col], crs=TARGET_CRS_GL).to_wkt()

# ── Step 7: Save ──────────────────────────────────────────────────────────
output_gl.to_file("glacial_lakes_1990_2015_timeseries.gpkg", driver="GPKG")
print(f"Saved: {len(output_gl)} lakes to glacial_lakes_1990_2015_timeseries.gpkg")
```
